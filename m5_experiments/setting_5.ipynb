```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from environment import *
from algorithms.maxcosd_box import MaxCOSD_Box_algorithm
from algorithms.parallelizer import Parallelizer
from algorithms.aim import AIM_algorithm

from simulator import Simulator
```

### Defining the environment
* Horizon: $T=1969$
* Number of runs: $N=1$
* Number of products: $n=3049$
* Demand distribution: $d_{t,i}\sim\text{Dataset}$
* Cost parameters: $(p_i/h_i) \propto 200$
* Feasible set: $\mathcal{Y}=\prod_{i=1}^n [0, \bar{y}_i]$ where $\bar{y}_i= 10\times(\text{optimal quantity for product }i)$
* Constants: $G = \sqrt{n}\max_{i\in[n]}\max\{h_i,p_i\}$

```python
m5_demands = pd.read_csv("m5_experiments/m5_demands.csv")
nb_samples = 1
horizon = m5_demands.period.max()
nb_products = len(m5_demands.product_id.unique())

costs_df = pd.read_csv("m5_experiments/m5_costs.csv")
holding_costs = costs_df["holding_cost"].values
penalty_costs = costs_df["penalty_cost"].values

demands = (m5_demands
    ["quantity"]
    .values
    .reshape(horizon+1,nb_products)
)
envs = [Environment_NonPerishable_Newsvendor(demands, holding_costs, penalty_costs)]
```

```python
y_max = np.zeros(nb_products)
for i in range(nb_products) :
    quantile = np.quantile(envs[0].demands[1:,i], penalty_costs[i]/(holding_costs[i]+penalty_costs[i]))
    y_max[i] = 10*quantile
```

### Defining the algorithms
* 3049 $\times$ 64 instances of AIM (one for each product and each value of $\gamma$ ranging from $10^{-5}$ to $10^1$)
* 3049 $\times$ 64 instances of MaxCOSD (one for each product and each value of $\gamma$ ranging from $10^{-5}$ to $10^1$)

```python
list_gamma = [10**exponent for exponent in np.linspace(-5,1,9*7+1)]

algs = [ Parallelizer([
            AIM_algorithm(np.zeros(1), np.zeros(1), y_max[product_id]*np.ones(1), gamma_value, np.max(np.maximum(holding_costs[product_id],penalty_costs[product_id])), holding_costs[product_id]*np.ones(1),penalty_costs[product_id]*np.ones(1)) 
            for product_id in range(nb_products)
        ], r"AIM $\gamma={:.3e}$".format(gamma_value)) for gamma_value in list_gamma]
algs += [ Parallelizer([
            MaxCOSD_Box_algorithm(np.zeros(1), np.zeros(1), y_max[product_id]*np.ones(1), gamma_value)
            for product_id in range(nb_products)
        ], r"MaxCOSD $\gamma={:.3e}$".format(gamma_value)) for gamma_value in list_gamma]

nb_algs = len(algs)
```

```python
horizons = np.unique(np.logspace(np.log10(1),np.log10(horizon),100).astype(int))

sim = Simulator(envs, nb_products, nb_samples, horizons, algs,
   lambda optimum: (optimum <= y_max).all(), holding_costs, penalty_costs)
cum_losses = sim.run()
```

### Figures
#### $R_T$ as a function of $\gamma$ in semilogx scale

```python
aim_mean_regret = np.array([np.mean(cum_losses[:,gamma_index,-1]-cum_losses[:,0,-1]) for gamma_index in range(1,len(list_gamma)+1)])
maxcosd_mean_cum_losses = np.array([np.mean(cum_losses[:,gamma_index,-1]-cum_losses[:,0,-1]) for gamma_index in range(1*len(list_gamma)+1,2*len(list_gamma)+1)])

plt.figure(figsize=(4.16,2.91))
plt.semilogx(list_gamma,aim_mean_regret, color="red", label="AIM",linestyle="--")
plt.semilogx(list_gamma,maxcosd_mean_cum_losses, color="blue", label="MaxCOSD")

plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.xlabel(r"$\gamma$")
plt.legend(loc="lower left")
plt.tight_layout()
plt.savefig("images/setting_5_gamma_logx.png", dpi=300)
plt.show()
```

#### $R_T$ as a function of $\gamma$ in loglog scale

```python
aim_mean_regret = np.array([np.mean(cum_losses[:,gamma_index,-1]-cum_losses[:,0,-1]) for gamma_index in range(1,len(list_gamma)+1)])
maxcosd_mean_cum_losses = np.array([np.mean(cum_losses[:,gamma_index,-1]-cum_losses[:,0,-1]) for gamma_index in range(1*len(list_gamma)+1,2*len(list_gamma)+1)])

plt.figure(figsize=(4.16,2.91))
plt.loglog(list_gamma,aim_mean_regret, color="red", label="AIM (Parallel)",linestyle="--")
plt.loglog(list_gamma,maxcosd_mean_cum_losses, color="blue", label="MaxCOSD (Parallel)")

plt.xlabel(r"$\gamma$")
plt.legend(loc="lower left")
plt.tight_layout()
plt.savefig("images/setting_5_gamma_loglog.png", dpi=300)
plt.show()
```

#### $R_T/T$ as a function of $T$ in loglog scale

```python
aim_gamma_index_opt = np.argmin(np.mean(cum_losses[:,1:len(list_gamma)+1,-1],axis=0))+1
maxcosd_gamma_index_opt = np.argmin(np.mean(cum_losses[:,1*len(list_gamma)+1:2*len(list_gamma)+1,-1],axis=0))+1*len(list_gamma)+1
colors = {aim_gamma_index_opt:"red", maxcosd_gamma_index_opt:"blue"}
linestyles = {aim_gamma_index_opt:"--", maxcosd_gamma_index_opt:"-"}

plt.figure(figsize=(5,3.5))

for i in [aim_gamma_index_opt, maxcosd_gamma_index_opt] :
    y = np.mean(cum_losses[:,i,:]-cum_losses[:,0,:],axis=0)
    y_q20 = np.quantile(cum_losses[:,i,:]-cum_losses[:,0,:],q=0.2, axis=0)
    y_q80 = np.quantile(cum_losses[:,i,:]-cum_losses[:,0,:],q=.8, axis=0)
    plt.loglog(horizons, y/horizons,label=algs[i-1], color=colors[i], linestyle=linestyles[i])
plt.loglog(horizons,10**5/np.sqrt(horizons),"-.",color="k",label=r"$\Theta(1/\sqrt{T})$")

plt.xlabel("Horizon (T)")
plt.legend(loc="lower left")
plt.tight_layout()
plt.savefig("images/setting_5_avgregret.png", dpi=300)
plt.show()
```
